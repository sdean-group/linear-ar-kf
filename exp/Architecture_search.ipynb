```python
import numpy as np
from define_system import define_system
from architecture_search import parameter_search
```

```python
n=4
p=2
m=3
rho_target=1.0
sys_seed = 0

A, B, C = define_system(n, p, m, rho_target, seed=sys_seed)
W = 0.05 * np.eye(n)
V = 0.1 * np.eye(m)

L = 10
H = 5
```

```python
search_range = np.arange(start=1, stop=11)
reps = 10
seeds = range(reps)
```

```python
num_best= {}
for i in search_range:
    num_best[i] = 0
all_losses = {}
```

```python
for i in range(reps):
    print(f'Running for repetition {i+1}')
    best_n, losses = parameter_search(A, B, C, W, V, L, H, 
                                      T=10000, 
                                      epochs=150,
                                      batch_size=64,
                                      lr=0.01,
                                      step_size=1,
                                      gamma=0.9,
                                      weight_decay=1e-3,
                                      train_seed=seeds[i], 
                                      torch_seed=100, 
                                      search_range=search_range)
    num_best[best_n] += 1
    all_losses[i] = losses
```

```python
avg_loss = {}
std_loss = {}

for i in search_range:
    losses = []
    for j in range(reps):
        train_curve = all_losses[j][i]
        losses.append(np.min(train_curve))

    avg_loss[i] = np.mean(losses)
    std_loss[i] = np.std(losses, ddof=1)
```

```python
ns = np.array(list(avg_loss.keys()))
avg_losses = np.array(list(avg_loss.values()))
std_losses = np.array(list(std_loss.values()))
```

```python
for i in search_range:
    print(f'Train loss of n={i} over {reps} seeds: mean = {avg_loss[i]:.4f} std = {std_loss[i]:.4f} ')
print(f'n with lowest mean loss: {min(avg_loss, key=avg_loss.get)}')
```
